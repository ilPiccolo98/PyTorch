```python
import torch

x = torch.rand(2, 2)
y = torch.rand(2, 2)

z = x + y
print(z)
z = torch.add(x, y)
print(z)
y.add_(x)
print(y)

#the same with subtraction, multiplication and division

```

```python
x = torch.rand(5, 3)
print(x[:])
print(x[:0])
print(x[1, :])
print(x[1, 1].item())
```

```python
x = torch.rand(4, 4)
print(x)
y = x.view(16)
print(y)
```

```python
import numpy as np

a = torch.ones(5)
print(a)
b = a.numpy()
print(type(b))

a.add_(1)
print(a)
print(b)
```

```python
a = np.ones(5)
print(a)
b = torch.from_numpy(a)
print(b)

a += 1
print(a)
print(b)
```

```python
if torch.cuda.is_available():
  device = torch.device("cuda")
  x = torch.ones(5, device=device)
  y = torch.ones(5)
  y = y.to(device)
  z = x + y
  z = z.to("cpu")
```
